```python
from scipy import stats
```

```python
x = [155,164,130,176,118,126,171,160,159,133]
```

```python
y = [38,40,54,81,45,56,76,72,61,44]
```

```python
slope, intercept, r, p, std_err = stats.linregress(x,y)
```

```python
def myfunc(x):
  return slope * x + intercept
```

```python
berat_badan = myfunc(155)
```

```python
print(berat_badan)
```
